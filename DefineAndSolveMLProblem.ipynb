### Handling Missing Values
1. Handle numerical columns by imputing missing values with median.
2. Handle string columns by feature engineering or dropping them. If we have time, we may use NLP techniques to extract relevant information from paragraph texts, but for now we use a simple approach.

```python
df_cleaned = df.copy()

print('\nMissing values per column:')
print(df_cleaned.isnull().sum())
```

```python
# Get columns with missing values
missing_cols = df_cleaned.isnull().sum()
missing_cols = missing_cols[missing_cols > 0]
print(missing_cols)
```

```python
# Impute missing values for host_response and _acceptance with median
response_median = df_cleaned['host_response_rate'].median()
acceptance_median = df_cleaned['host_acceptance_rate'].median()

df_cleaned['host_response_rate'].fillna(value=response_median, inplace=True)
df_cleaned['host_acceptance_rate'].fillna(value=acceptance_median, inplace=True)
```

To deal with missing values in `bedrooms` and `beds` columns, we can group by `room_type` and impute the median value of that group for missing values. This accounts for the fact that different housing types have different number of bedrooms and beds.

```python
bedroom_medians = df_cleaned.groupby('room_type')['bedrooms'].median()
bed_medians = df_cleaned.groupby('room_type')['beds'].median()

df_cleaned['bedrooms'] = df_cleaned['bedrooms'].fillna(df_cleaned['room_type'].map(bedroom_medians))
df_cleaned['beds'] = df_cleaned['beds'].fillna(df_cleaned['room_type'].map(bed_medians))
```

#### Handling Missing Text Columns

We will drop these text columns, but before doing so we can extract some useful data: missingness (whether the example has a value for that field or not) and the length of the text.

```python
# Name: Too many unique values, create column for length of the name to capture some info
df_cleaned['name_length'] = df_cleaned['name'].fillna('').apply(len)

# Description: capture missingness and length
df_cleaned['description_length'] = df_cleaned['description'].fillna('').apply(len)
df_cleaned['description_missing'] = df_cleaned['description'].isnull().astype(int)

# Neighborhood Overview: capture missingness
df_cleaned['neighborhood_overview_missing'] = df_cleaned['neighborhood_overview'].isnull().astype(int)

# Host Location: we can drop this column as it carries little signal

# Host_about: capture missingness and length
df_cleaned['host_about_length'] = df['host_about'].fillna('').apply(len)
df_cleaned['host_about_missing'] = df['host_about'].isnull().astype(int)
```

```python
# Drop original columns
df_cleaned.drop(['name', 'description', 'neighborhood_overview', 'host_location', 'host_about'], axis=1, inplace=True)
df_cleaned.isnull().sum().sum() # 0 missing values left
```

### Handling Outliers

For the price column, we can handle outliers by log-transforming the values so the distribution resembles a bell-shaped curve, then winsorize the outliers that are left.

We can follow a similar approach for the min number of nights that the listing can be booked for. Some of the listings have large values for `minimum_nights` that may be a mistake, or otherwise not representative of the dataset.

```python
import scipy.stats as stats

# Create Log-transformed price column
df_cleaned['price_log'] = np.log1p(df['price'])
# Winsorize top and bottom 5% of price points
df_cleaned['price_log'] = stats.mstats.winsorize(df_cleaned['price_log'], limits=[0.05, 0.05])

diff = (df_cleaned['price_log'] - np.log1p(df['price'])).nunique()
print(f'Price Winsorized Count: {diff}')
```

```python
sns.boxplot(df['minimum_nights'])
plt.title("Boxplot of Minimum Nights")
plt.show()
```

```python
df_cleaned['minimum_nights'] = stats.mstats.winsorize(df['minimum_nights'], limits=[0.05, 0.05])
diff = (df_cleaned['minimum_nights'] - df['minimum_nights']).nunique()
print(f'Winsorized Count: {diff}')
```

### Feature Engineering

For the remaining text columns, we can engineer features to extract information such as length of text or relevant keywords, or simply drop them. For categorical data `neighbourhood_group_cleansed` and `room_type`, we can use one-hot-encoding. For `amenities`, which is a string representation of a list of strings separated by commas, we can include a column for the length of the amenities list. Finally, we can consider dropping the `host_name` column, as the host's name likely does not impact the listing's rating.

```python
df_transform = df_cleaned.copy()
df_transform.select_dtypes(include=['object', 'string']).columns
```

```python
# One-hot encode neighbourhood_group_cleansed and room_type
df_transform = pd.get_dummies(df_transform, columns=['neighbourhood_group_cleansed', 'room_type'], drop_first=True)

# Count numbers of amenities (separated by commas)
df_transform['amenities_count'] = df_transform['amenities'].apply(lambda x: x.count(',') + 1 if x else 0)

# Drop remaining text columns
df_transform.drop(['host_name','amenities'], axis=1, inplace=True)
```

### Split the Data for Test and Train

```python
# YOUR CODE HERE

# Remove data leakage features
to_drop = ['review_scores_cleanliness',
 'review_scores_checkin',
 'review_scores_communication',
 'review_scores_location',
 'review_scores_value']
df_transform.drop(columns=to_drop, inplace=True)

df_transform.dropna(subset=['review_scores_rating'], inplace=True)

# X and y
X = df_transform.drop(columns=['review_scores_rating'])
y = df_transform['review_scores_rating']

# train test split
X_temp, X_test, y_temp, y_test = train_test_split(X, y, test_size=0.2)
X_train, X_val, y_train, y_val = train_test_split(X_temp, y_temp, test_size=0.25)
```

### Models

#### KNN

```python
# Build kNN pipeline with feature scaling

knn_pipeline = Pipeline([
    ('scaler', StandardScaler()),
    ('knn', KNeighborsRegressor())
])

knn_pipeline.fit(X_train, y_train)

y_pred_val_knn = knn_pipeline.predict(X_val)

mse_knn = mean_squared_error(y_val, y_pred_val_knn)
rmse_knn = np.sqrt(mse_knn)
r2_knn = r2_score(y_val, y_pred_val_knn)

print(f"K-Nearest Neighbors Regressor Model Evaluation:")
print(f"  Mean Squared Error (MSE): {mse_knn:.4f}")
print(f"  Root Mean Squared Error (RMSE): {rmse_knn:.4f}")
```

#### Decision Tree Regressor

```python
dt_model = DecisionTreeRegressor()
dt_model.fit(X_train, y_train)

y_pred_val_dt = dt_model.predict(X_val)

mse_dt = mean_squared_error(y_val, y_pred_val_dt)
rmse_dt = np.sqrt(mse_dt)
r2_dt = r2_score(y_val, y_pred_val_dt)

print(f"Decision Tree Regressor Model Evaluation:")
print(f"  Mean Squared Error (MSE): {mse_dt:.4f}")
print(f"  Root Mean Squared Error (RMSE): {rmse_dt:.4f}")
```

#### Random Forest Regressor

```python
rf_model = RandomForestRegressor()
rf_model.fit(X_train, y_train)

y_pred_val_rf = rf_model.predict(X_val)

mse_rf = mean_squared_error(y_val, y_pred_val_rf)
rmse_rf = np.sqrt(mse_rf)
r2_rf = r2_score(y_val, y_pred_val_rf)

print(f"Random Forest Regressor Model Evaluation:")
print(f"  Mean Squared Error (MSE): {mse_rf:.4f}")
print(f"  Root Mean Squared Error (RMSE): {rmse_rf:.4f}")
```

#### Grid Search for Gradient Boosting Regressor

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import GradientBoostingRegressor

# Define the parameter grid for Gradient Boosting Regressor
param_grid_gbr = {
    'n_estimators': [100, 200, 300],
    'learning_rate': [0.01, 0.1, 0.2],
    'max_depth': [3, 4, 5],
    'min_samples_split': [2, 5],
    'min_samples_leaf': [1, 2]
}

# Initialize GridSearchCV
grid_search_gbr = GridSearchCV(estimator=GradientBoostingRegressor(random_state=42),
                              param_grid=param_grid_gbr,
                              scoring='neg_mean_squared_error',
                              cv=3,
                              n_jobs=-1)

# Fit GridSearchCV to the training data
grid_search_gbr.fit(X_train, y_train)

# Get the best parameters and best estimator
best_params_gbr = grid_search_gbr.best_params_
best_gbr_model = grid_search_gbr.best_estimator_

print(f"Best parameters for Gradient Boosting Regressor: {best_params_gbr}")

# Evaluate the best model on the validation set
y_pred_val_best_gbr = best_gbr_model.predict(X_val)

mse_best_gbr = mean_squared_error(y_val, y_pred_val_best_gbr)
rmse_best_gbr = np.sqrt(mse_best_gbr)
r2_best_gbr = r2_score(y_val, y_pred_val_best_gbr)

print(f"\nBest Gradient Boosting Regressor Model Evaluation on Validation Set:")
print(f"  Mean Squared Error (MSE): {mse_best_gbr:.4f}")
print(f"  Root Mean Squared Error (RMSE): {rmse_best_gbr:.4f}")
print(f"  R-squared (R2): {r2_best_gbr:.4f}")
```
